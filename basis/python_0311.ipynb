### 벡터의 합 - 두 벡터를 이웃하는 변으로 만든 평행 사변형의 대각선 벡터

```python
a = np.array([1, 2])
b = np.array([2, 1])
c = a + b
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=c, xytext=(0, 0), arrowprops=black)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.plot(c[0], c[1], 'ro', ms=10)
plt.plot([a[0], c[0]], [a[1], c[1]], 'k--')
plt.plot([b[0], c[0]], [b[1], c[1]], 'k--')
plt.text(0.35, 1.15, "$a$")
plt.text(1.15, 0.25, "$b$")
plt.text(1.25, 1.45, "$c$")
plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.show()
```

### 벡터의 합 - 두번째의 벡터를 첫번째 벡터의 끝점으로 평행이동한 후 원점에서부터 평행 이동한 벡터의 끝점까지의 벡터와 동일


```python
a = np.array([1, 2])
b = np.array([2, 1])
c = a + b
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=c, xytext=a, arrowprops=gray)
plt.annotate('', xy=c, xytext=(0, 0), arrowprops=black)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(c[0], c[1], 'ro', ms=10)
plt.text(0.35, 1.15, "$a$")
plt.text(1.45, 2.45, "$b$")
plt.text(1.25, 1.45, "$c$")
plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.show()
```

### 벡터의 선형 조합 - 여러 벡터에 스칼라 값을 곱해서 더한 것

## 좌표 변환

### rank - 선형 독립인 벡터의 최대 개수

```python
#앞의 데이터를 어떻게 조작해도 뒤의 데이터를 만들 수 없음 - rank 2
ar = np.array([[1, 2], [3, 3]])
print(np.linalg.matrix_rank(ar))

#앞의 2개의 선형 조합으로 세번째 데이터를 만들 수 있음 - rank 2
ar = np.array([[1, 2], [3, 3], [5, 7]])
print(np.linalg.matrix_rank(ar))
```

### 고유 값과 고유 벡터

```python
B = np.array([[2, 1], [1, -1]])
w2, v2 = np.linalg.eig(B)
print(w2)#고유값
print(v2)#오른쪽 고유 벡터
```

### 특잇값 분해

```python
from numpy.linalg import svd

A = np.array([[3, -1], [1, 3], [1, 1]])
U, S, VT = svd(A)
print(U)
#축소된 형태로 표시
print(S)
print(VT)
```

```python
#축소된 행렬을 복원
print(np.diag(S, 1)[:, 1:])
```

```python
#분해된 행렬을 이용해서 원래의 행렬을 복원
print(U @ np.diag(S, 1)[:, 1:] @ VT)
#분해된 행렬을 가지고 원래의 행렬을 만들 때는 제대로 복원이 되지 않을 수 있음
```

### 붓꽃의 성분 간의 종속성 확인

```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data[:10, :2] # 10개 데이터의 꽃받침의 길이 와 꽃받침 폭만 선택
print(X)
```

```python
#읽어온 데이터를 시각화
plt.figure(figsize=(8,8))

plt.plot(X.T, 'o:')
plt.xticks(range(2), ['꽃받침 길이', '꽃받침 폭'])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title("붓꽃 크기 특성")
plt.legend(['[표본 {}]'.format(i+1) for i in range(10)])
plt.show()
```

```python
plt.figure(figsize = (8,8))
ax = sns.scatterplot(0, 1, 
                     data=pd.DataFrame(X), s=100, color='.2', marker='s')
plt.xlabel("꽃받침 길이")
plt.ylabel('꽃받침 폭')
plt.title('붓꽃 크기 특성(2차원 표시)')
plt.show()
```
