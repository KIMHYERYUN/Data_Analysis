## csv 파일을 읽어서 그래프로 출력하기

```python
#데이터 읽기

#한글이 깨지면 encoding을 추가
f = open('./data/mokpo.csv', encoding='ms949')
data = csv.reader(f)

#첫번째 줄은 열의 제목이므로 제외
next(data)

#필요한 데이터 추출
#매년 8월 18일의 최고 기온 과 최저 기온만 추출

#년도 와 최고기온 과 최저기온을 저장할 list
year = []
high = []
low = []

for row in data:
    #첫번째 열에 해당하는 데이터 가져오기 - 날짜
    date = row[0]
    #.을 기준으로 분할 - x[0]은 연도, x[1]은 월, x[2]은 일
    x = date.split('.')
    if x[1] == '8' and x[2] == '18':
        #년도를 저장
        year.append(int(x[0]))
        #최고 기온을 저장
        high.append(float(row[-1]))
        #최저 기온을 저장
        low.append(float(row[-2]))
        
#제대로 저장되었는지 샘플을 출력
for idx in range(0, 5, 1):
    print(year[idx], ":", high[idx], ":", low[idx])
    
#그래프 영역 설정
plt.figure(figsize=(10, 4))

#그래프 생성
plt.plot(high, label='최고 기온', lw=3, 
         color='r', linestyle='-', marker='s')
plt.plot(low, label='최저 기온', lw=3, 
         color='b', linestyle='-', marker='s')

#그래프 옵션 설정
#숫자의 범위 조절
plt.ylim(-20, 40)
#x 축 눈금 설정
plt.xticks(range(0, len(year), 1), year,rotation='vertical')
#범례 출력
plt.legend()

#그래프 출력
plt.show()
```

```python
# 입력받은 도시의 성별 나이별 인구 분포를 확인하기
f = open('./data/pop.csv', encoding='ms949')
data = csv.reader(f)

#도시 이름 입력받기
city = input('도시 이름을 공백없이 입력하시오!')

#입력받은 도시의 남자 인구 와 여자 인구를 저장하기 위한 list
man = []
woman = []

for row in data:
    #in 연산자는 포함된
    if city in row[0]:
        for i in range(24, 45, 1):
            #남자 데이터만 정수로 변환해서 저장
            man.append(int(row[i].replace(',', '')))
            woman.append(int(row[i+23].replace(',', '')))
            
plt.figure(figsize=(10, 4))
plt.plot(man, label='남자')
plt.plot(woman, label='여자')
plt.xticks(range(0,21), range(0,105,5), rotation='vertical')
plt.legend()
plt.show()
```

### 산포도 그리기

```python
# 입력받은 도시의 성별 나이별 인구 분포를 확인하기
f = open('./data/pop.csv', encoding='ms949')
data = csv.reader(f)

#도시 이름 입력받기
city = input('도시 이름을 공백없이 입력하시오!')

#입력받은 도시의 남자 인구 와 여자 인구를 저장하기 위한 list
man = []
woman = []

for row in data:
    #in 연산자는 포함된
    if city in row[0]:
        for i in range(24, 45, 1):
            #남자 데이터만 정수로 변환해서 저장
            man.append(int(row[i].replace(',', '')))
            woman.append(int(row[i+23].replace(',', '')))
            
plt.figure(figsize = (5, 2))

#산포도 그리기
plt.scatter(x=man, y=woman, marker='3', c=range(0, 21, 1), 
           alpha=0.5, cmap='jet')
plt.colorbar()
plt.title(city + ' 성별 인구 분포')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')

plt.show()
```

## 히스토그램

```python
#data/korea.csv 파일 읽기

#이름 과 점수를 저장할 list
names = []
scores = []

#파일을 개방
f = open('./data/korea.csv', encoding='ms949')
data = csv.reader(f)

#첫번째 행이 열의 제목이므로 넘어가기
next(data)

#각 줄을 읽어서 첫번째 데이터는 names 에 두번째 데이터는 scores 에 추가
for row in data:
    names.append(row[0])
    scores.append(row[1])

#scores 의 모든 데이터를 정수로 변환
#map : 컬렉션의 데이터에 함수를 적용해서 데이터의 모임을 리턴
scores = list(map(int, scores))
    
#히스토그램 그리기
plt.figure()
#히스토그램을 그릴 때는 (빈도수 파악) 범주형 데이터가 아닌 경우는
#구간화(bining)를 하는 것이 좋습니다.
plt.hist(scores, bins=3)
plt.xticks(range(0,100,25), labels=['', '하', '중', '상'])

plt.title('점수 빈도')
plt.xlabel('점수')
plt.ylabel('빈도수')
plt.show()
```

## list 와 의 작업 시간 비교

```python
import numpy as np
import datetime

#list 생성 - 0부터 9999999 까지
li = list(range(0, 10000000, 1))

#현재 시간을 가져오기
start = datetime.datetime.now()
print("리스트 작업 시작 시간:", start)

#모든 데이터에 10을 곱해서 저장
for i in li:
    i * 10

end = datetime.datetime.now()
print("리스트 작업 종료 시간:", end)

ar = np.arange(0, 10000000, 1)

#현재 시간을 가져오기
start = datetime.datetime.now()
print("배열 작업 시작 시간:", start)

#배열의 모든 요소에게 10을 곱해줍니다.
ar * 10

end = datetime.datetime.now()
print("배열 작업 종료 시간:", end)
```

## ndarray 의 생성 과 정보 확인 - 실제 생성은 학습할 때만 이렇게 합니다.

```python
import numpy as np

#일차원 배열 생성
ar = np.array([1, 2, 3])

#ar 의 자료형 확인
print(type(ar))

#ar 의 요소의 자료형 확인
print(ar.dtype)

print(ar.ndim) # 1차원
print(ar.shape) #1차원인데 데이터는 3개

ar = np.array([[1, 2, 3], [4, 5, 6]]) # 2차원 배열
print(ar.ndim) #2차원
print(ar.shape) #1차원인데 데이터는 3개
```

## numpy 의 함수를 이용한 ndarray 생성

## 특수 행렬

```python
import numpy as np
#1로 10를 채운 배열을 생성
ar1 = np.ones(10)
print(ar1)

#1로 채워진 5*5 배열 생성
ar2 = np.ones((5, 5))
print(ar2)

#2*2 정방 행렬을 만들고 대각선 방향으로 1을 채운 행렬 생성
ar3 = np.eye(2)
print(ar3)

#1의 위치를 수정
ar4 = np.eye(3, k=1)
print(ar4)

#대각선에서 데이터를 추출해서 새로운 1차원 배열 생성
br = np.diag(ar4, k=1)
print(br)
```

## 요소의 데이터 타입을 변경

```python
import numpy as np

#type을 설정해서 생성
ar = np.array([1,2,'3'], dtype=np.int32)
print(ar.dtype)

#type을 float32 로 변경
ar = ar.astype(np.float32)
print(ar.dtype)
```

## 행렬의 구조 변경 - reshape, flattern

```python
import numpy as np

#1차원 배열을 생성
ar = np.arange(10)
print(ar)

#2차원 배열로 변환
br = ar.reshape(2,5)
print(br)

#2차원 배열을 1차원 배열로 변환
print(br.reshape(-1))
print(br.flatten())


#1차원 배열을 생성
ar = np.arange(20)
#1차원 배열을 3차원으로 변환
print(ar.reshape(2, 5, 2))
#마지막 차원의 개수는 -1로 설정하면 알아서 분할해서 생성해 줍니다.
print(ar.reshape(2, 5, -1))
```
