Series

Series를 직접 생성하고 속성 확인

```python
#pandas라는 모듈을 pd라는 이름을 가져온 것
import pandas as pd

#pandas패키지의 Series 모듈을 현재 애플리케이션의 메모리로 가져와서 사용
#Series 앞에는 pandas.를 붙일 필요가 없음
from pandas import Series

#list를 이용해서 Series를 생성
#index를 설정하지 않아서 index는 0부터 시작하는 시퀀스
price = Series([4000, 3000, 2000, 1000])
#데이터 확인
print(price) #인덱스,값,dtype 출력

#인덱스 확인
print(price.index)

#데이터를 확인
print(price.values)
print(type(price.values))  #자료형은 numpy의 ndarray

#인덱스 변경
price.index = ['apple', 'banana', 'orange', 'mango']
print(price)
```

Series의 인덱싱

```python
#pandas라는 모듈을 pd라는 이름을 가져온 것
import pandas as pd

#pandas패키지의 Series 모듈을 현재 애플리케이션의 메모리로 가져와서 사용
#Series 앞에는 pandas.를 붙일 필요가 없음
from pandas import Series

sourcedata = {'kim':29, 'park':35, 'lee':15}
#dict를 이용해서 Series를 생성
#dict를 가지고 생성하면 key가 인덱슥가 되고 value는 value가 된다.
data = Series(sourcedata)
print(data)

#데이터 1개 가져오기
print(data[0]) #0번 위치
print(data['kim']) #kim이라는 인덱스를 가진 데이터
print()

#범위설정
print(data[0:2]) #0번에서 2번 앞까지
print(data['kim':'lee']) #kim부터 lee까지
print()

#Fancy Indexing - 복제가 발생
print(data[['kim', 'lee']])
print()

#Boolean Indexing - True인 데이터만 필터링
print(data[[True, False, False]])
```

Series 연산

```python
#pandas라는 모듈을 pd라는 이름을 가져온 것
import pandas as pd

#pandas패키지의 Series 모듈을 현재 애플리케이션의 메모리로 가져와서 사용
#Series 앞에는 pandas.를 붙일 필요가 없음
from pandas import Series

#하나의 list는 key로 다른 하나의 list는 value로 해서 dict를 생성
keys = ['kim', 'park', 'hur', 'jung']
values = [30, 19, 35, 23]
dict1 = dict(zip(keys, values))
print(dict1)

keys = ['kim', 'park', 'hur', 'choi']
values = [30, None, 35, 23]
dict2 = dict(zip(keys, values))
print(dict2)

s1 = Series(dict1)
s2 = Series(dict2)
print(s1)
print(s2)

print(s1+s2) #한 쪽에만 데이터 존재하거나 하나의 데이터가 None일 경우, NaN
```
