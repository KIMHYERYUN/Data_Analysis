###pipe함수 - 함수에 함수를 적용하는 것이 가능

```python
#하나의 값을 받아서 결과를 리턴하는 함수
def missing_value(x):
    return x.isnull()

#위 함수의 결과를 집계해서 개수를 리턴
def missing_count(x):
    return missing_value(x).sum()

#위 함수의 결과를 이용해서 집계
def total_missing_count(x):
    return missing_count(x).sum()

print(df.pipe(missing_value))
print(df.pipe(missing_count))
print(df.pipe(total_missing_count))
```

#필터링 - 원하는 조건에 맞는 데이터만 추출

```python
import numpy as np
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.info()

#age가 30이상인 데이터만 추출
titanic.loc[titanic['age']>=30]

#age가 30~39인 데이터만 추출
titanic.loc[(titanic['age']>=30) & (titanic['age'] <= 39)]

#age가 30~39인 데이터 중 여자만 추출
titanic.loc[(titanic['age']>=30) & (titanic['age'] <= 39) & (titanic['sex'] =='female')]
```

```python
import numpy as np
import pandas as pd
import seaborn as sns

df1 = pd.DataFrame({'a':['1', '2', '3', '4'], 
                   'b':['a', 'b', 'c', 'd']})
df2 = pd.DataFrame({'a':['11', '22', '33', '44'], 
                   'b':['aa', 'bb', 'cc', 'dd']})

#위 아래로 결합
print(pd.concat([df1, df2]))

#좌우로 결합
print(pd.concat([df1, df2], axis=1))
```

```python
#기본적으로 outer join
df1 = pd.DataFrame({'a':['1', '2', '3', '4'], 
                   'b':['a', 'b', 'c', 'd']},
                  index = [0,1,2,3])
df2 = pd.DataFrame({'a':['11', '22', '33', '44'], 
                   'b':['aa', 'bb', 'cc', 'dd']},
                  index = [2,3,4,5])

#위 아래로 결합
print(pd.concat([df1, df2]))

#좌우로 결합 -outer
print(pd.concat([df1, df2], axis=1))

#좌우로 결합 -inner
print(pd.concat([df1, df2], axis=1, join='inner'))
```

merge - join과 유사

```python
#데이터 가져오기
price = pd.read_excel('c://programming/python/pandasdata/stock price.xlsx')
price.info()

valuation = pd.read_excel('c://programming/python/pandasdata/stock valuation.xlsx')
valuation.info()

#양쪽모두에 존재하는 데이터만 결합 - id라는 공통 열을 가지고 join
merge_inner = pd.merge(price, valuation)
print(merge_inner)

#모든 데이터 결합 - id라는 공통 열을 가지고 join
merge_outer = pd.merge(price, valuation, how='outer', on='id')
print(merge_outer)

#서로 다른 열 이름을 가지고 join
merge_outer = pd.merge(price, valuation, how='outer', left_on='stock_name', right_on='name')
print(merge_outer)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.info()

#age, sex, class, fare, survived 열만 추출
df = titanic[['age', 'sex', 'class', 'fare', 'survived']]
print(df.head())

#class별로 그룹화
grouped = df.groupby(['class'])
print(grouped)

#그룹 객체를 iterator를 이용해서 출력
for key, group in grouped:
    print('key:', key)
    print('   count:', len(group))
    print(group.head())
    print()

#그룹별 평균
average = grouped.mean()
print(average)

#그룹별 선택
roup1 = grouped.get_group('First')
print(group1.head())

#2개의 항목으로 그룹화
group_two = df.groupby(['class', 'sex'])
for key, group in group_two:
    print('key:', key)
    print('   count:', len(group))
    print(group.head())
    print()
```

그룹별 집계

```python
#seaborn의 titanic 데이터를 가져와서 class별로 그룹화

import numpy as np
import pandas as pd
import seaborn as sns


titanic = sns.load_dataset('titanic')

grouped = titanic.groupby(['class'])

for key, group in grouped:
    print(key)
    print(group.head())
    
#모든 열의 표준편차를 구해서 출력
grouped.std()

#특정 열의 표준편차를 구해서 출력
grouped.fare.std()
grouped['fare'].std()

#사용자 정의 함수를 적용 - 최대값에서 최소값을 뺀값을리턴
def min_max(x):
    return x.max() - x.min()

grouped.agg(min_max)

#여러 개 함수 적용
grouped.agg(['min', 'max', 'sum'])
```

변환

```python
#z-score를 구해주는 함수
#평균을 뺀값을 표준편차로 나누어서 리턴하는 함수
def z_score(x):
    return (x - x.mean()) / x.std()

age_zscore = grouped.age.transform(z_score)
print(age_zscore)
```

필터링 - RDBMS의 SQL에서 Having

```python
#필터링을 할 때는 매개변수가 1개이고 bool을 리턴하는 함수를 이용


#데이터 개수가 200 이상인 그룹만 추출
def over_200(x):
    return len(x) >= 200
#Second 그룹 제외
filtering_group = grouped.filter(over_200)
print(filtering_group.head())

#람다 이용 - 람다가 한 줄 짜리 이름없는 함수
filtering_group = grouped.filter(lambda x:len(x) >= 200)
print(filtering_group.head())
```

```python
#seaborn의 titanic 데이터를 가져와서 class별로 그룹화
import numpy as np
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic[['age', 'sex', 'class', 'fare', 'survived']]
print(df)

pivot_1 = pd.pivot_table(df, index='class', columns = 'sex', values = 'age', aggfunc = 'mean')
print(pivot_1)
```
