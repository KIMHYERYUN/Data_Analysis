```python
#데이터 읽기
import csv
#한글이 깨지면 encoding 추가
f = open('C://programming/python/data/visualization.csv', encoding='ms949')
data = csv.reader(f)

#첫번째줄은 열의 제목으로 제외
next(data)

#필요한 데이터 추출
#매년 8월 18일의 최고 기온과 최저 기온만 추출

#년도와 최고기온과 최저기온을 저장할 list
year = []
high = []
low = []

for row in data:
    #첫번째 열에 해당하는 데이터 가져오기 - 날짜
    date = row[0]
    #.를 기준으로 분할
    x = date.split('/')
    #print(x[1])
    
    if x[0] == '1' and x[1] =='15':
        #년도를 저장
        year.append(int(x[2]))
        #최고기온을 저장
        high.append(float(row[-1]))
        #최저기온을 저장
        low.append(float(row[-2]))
        
#제대로 저장되었는지 샘플 출력
for idx in range(0, 5, 1):
    print(year[idx], ":", high[idx], ":", low[idx])
                                                  
#그래프 영역 설정
plt.figure(figsize=(10, 4))

#그래프 생성
plt.plot(high, label='최고기온', lw=3, color='r', linestyle='-', marker='s')
plt.plot(high, label='최저기온', lw=3, color='b', linestyle='-', marker='s')
                                                  
#그래프 옵션 설정
#숫자 범위 조절
plt.ylim(-20, 40)
#x축 눈금 설정
plt.xticks(range(0, len(year), 1), year, rotation='vertical')
#범례 출력
plt.legend()
#그래프 출력
plt.show()
```

```python
# 입력받은 도시의 성별 나이별 인구 분포를 확인하기
f = open('c://programming/python/data/pop.csv', encoding='ms949')
data = csv.reader(f)

#도시 이름 입력받기
city = input('도시 이름을 공백없이 입력하시오!')

#입력받은 도시의 남자 인구 와 여자 인구를 저장하기 위한 list
man = []
woman = []

for row in data:
    #in 연산자는 포함된
    if city in row[0]:
        for i in range(24, 45, 1):
            #남자 데이터만 정수로 변환해서 저장
            man.append(int(row[i].replace(',', '')))
            woman.append(int(row[i+23].replace(',', '')))
            
plt.figure(figsize=(10, 4))
plt.plot(man, label='남자')
plt.plot(woman, label='여자')
plt.xticks(range(0,21), range(0,105,5), rotation='vertical')
plt.legend()
plt.show()
```

```python
# 입력받은 도시의 성별 나이별 인구 분포를 확인하기
f = open('c://programming/python/data/pop.csv', encoding='ms949')
data = csv.reader(f)

#도시 이름 입력받기
city = input('도시 이름을 공백없이 입력하시오!')

#입력받은 도시의 남자 인구 와 여자 인구를 저장하기 위한 list
man = []
woman = []

for row in data:
    #in 연산자는 포함된
    if city in row[0]:
        for i in range(24, 45, 1):
            #남자 데이터만 정수로 변환해서 저장
            man.append(int(row[i].replace(',', '')))
            woman.append(int(row[i+23].replace(',', '')))
            
plt.figure(figsize = (5, 2))

#산포도 그리기
plt.scatter(x=man, y=woman, marker='3', c=range(0, 21, 1), 
           alpha=0.5, cmap='jet')
plt.colorbar()
plt.title(city + ' 성별 인구 분포')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')

plt.show()
```
