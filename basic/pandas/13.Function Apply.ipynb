## 데이터에 함수 적용(apply)

```python
#키 뒤에 cm 붙이는 함수 만들기 - apply(함수명)
def add_cm(height):
    return str(height) + "cm"

df['키'] = df['키'].apply(add_cm)
df
```

```python
#첫번째 글자 대문자 변경 함수 만들기
def capitalize(lang):
    if pd.notnull(lang): #NaN이 아닌지
        return lang.capitalize() #첫 글자는 대문자, 나머지는 소문자
    return lang
df['특기'] = df['특기'].apply(capitalize)
df
```
