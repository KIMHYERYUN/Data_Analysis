# 14.그룹화
동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.groupby('학교')
```

```python
#학교 중 '덕원여고' 그룹화하여 데이터 보여주기
df.groupby('학교').get_group('덕원여고')
```

```python
#학교로 그룹화 - 평균내기
df.groupby('학교').mean()
```

```python
#학교로 그룹화 - 크기(갯수) 정보
df.groupby('학교').size()
```

```python
#학교로 그룹화 - 크기(갯수) 정보 중 일부만
df.groupby('학교').size()['덕원여고']
```

```python
#그룹화 - 단일 열의 데이터
#학교로 그룹화 - 키 열 - 평균내기
df.groupby('학교')['키'].mean()
```

```python
#그룹화 - 여러 열의 데이터
#학교로 그룹화 - 국,영,수 열 - 평균내기
df.groupby('학교')[['국어', '수학', '영어']].mean()
```

```python
#학년 column 추가
df['학년'] = [3, 1, 2, 2, 1, 3, 1]
df
```

```python
#여러 개의 열로 그룹화 - 평균내기
df.groupby(['학교', '학년']).mean()
```

```python
#그룹화 - 평균 - 키 기준  오름/내림차순
df.groupby('학년').mean().sort_values('키', ascending=False)
```

```python
#그룹화 - 특정 열에 대한 값 구하기
df.groupby('학교')['특기'].count()
```

```python
#그룹화 - 여러 열에 대한 값 구하기
df.groupby('학교')[['이름', '특기']].count()
```

```python
#그룹화 - 특정 열에 대한 값 구하기
df.groupby('학교')['학년'].value_counts()
```

```python
#그룹화 - 특정 열에 대한 값의 일부분 데이터 구하기
df.groupby('학교')['학년'].value_counts().loc['덕원여고']
```

```python
#그룹화 - 특정 열에 대한 값의 일부분 데이터 구하기 - 퍼센트
df.groupby('학교')['학년'].value_counts(normalize=True).loc['덕원여고']
```
