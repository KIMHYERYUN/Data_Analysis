# 12.데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

# Column 수정

```python
#값을 바꾸기 - 동기화 x
df['학교'].replace({'덕원여고':'덕원여중'})
```

```python
#여러 값을 바꾸기 - 동기화 x
df['학교'].replace({'덕원여고':'덕원여중', '광영여고':'광영고'})
```

```python
#여러 값을 바꾸기 - 동기화 o
df['학교'].replace({'덕원여고':'덕원여중', '광영여고':'광영고'}, inplace=True)
df
```

```python
#특정 열에 대한 값 수정하기 - 바꾸기
df['특기'] = df['특기'].str.upper()
df
```

```python
#특정 열에 대한 값 수정하기 - 추가하기
df['이름'] = df['이름'] + '학생'
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['수학'] + df['영어'] + df['과학'] + df['사회']
df
```

```python
#'결과' 열 추가하고 조건에 맞게 값 입력
df['결과'] = 'Fail'
#조건에 맞는 열 출력
df.loc[df['총합']>400, '결과'] = 'Pass'
df
```

## Column 삭제

```python
#일부 열 삭제 - 동기화 x
df.drop(columns=['총합'])
```

```python
#여러 열 삭제 - 동기화 ㅇ
df.drop(columns=['국어', '영어', '수학'], inplace=True) 
```

## Row 삭제

```python
#index 활용
df.drop(index='4번')
```

```python
#조건에 맞는 행 지우기
#-조건 해당 index 구하기
filt = df['수학'] < 80
df[filt].index
```

```python
#해당 index 활용하여 삭제
df.drop(df[filt].index)
```

## Row 추가

```python
df
```

```python
#새로운 row 추가
df.loc['8번'] = ['shim학생', '광영고', 183, 80, 70, 99, 86, 75, 'KOTLINE', 420, 'Fail']
df
```

## Cell 수정

```python
#한개의 데이터 수정
df.loc['4번', '특기'] = 'C'
df
```

```python
#여러 데이터 수정
df.loc['3번', ['학교', '특기']] = ['덕원여고', 'C+']
df
```

## Column 순서 변경

```python
#columns의 리스트 출력
cols = list(df.columns)
cols
```

```python
df[cols[0:-1]]
```

```python
#맨 뒤의 열을 맨 앞으로 이동 후 나머지 열 가져오기
#-cols[-1] -> series로 하나의 값으로 DataFrame 형식으로 [] 한번 더 감싸줘야 함 
df = df[[cols[-1]]+cols[0:-1]]
df
```
