```python
import pandas as pd
data = {
    '이름' : ['kim', 'hur', 'park', 'jung', 'lee', 'choi', 'han'],
    '학교' : ['덕원여고', '덕원여고', '덕원여고', '화곡고', '광영여고', '덕원여고', '화곡고'],
    '키' : [167, 165, 180, 172, 176, 161, 159],
    '국어' : [85, 90, 70, 89, 100, 61, 50],
    '수학' : [95, 81, 73, 86, 91, 40, 80],
    '영어' : [90, 95, 89, 61, 70, 80, 86],
    '과학' : [88, 79, 60, 95, 86, 54, 75],
    '사회' : [80, 93, 91, 80, 94, 79, 81],
    '특기' : ['python', 'java', 'c', 'c', 'java', 'javascript', 'python']
}
df=pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번'])
df
```

```python
df.index
```

Index 이름 설정

```python
df.index.name = '지원번호'
df
```

Index 초기화

```python
df.reset_index()
```

```python
#기존 사용한 '지원번호' index 삭제
df.reset_index(drop=True)
```

```python
#실제 확인해보면 미반영
df
```

```python
#실제 반영하기
df.reset_index(drop=True, inplace=True)
```

```python
df
```

## Index 설정
지정한 column으로 Index 설정

```python
#'이름'으로 Index 설정 - 미반영 상태
df.set_index('이름')
```

```python
#'이름'으로 Index 설정 - 반영 상태
df.set_index('이름', inplace=True)
df
```

## Index 정렬
Index를 기준으로 오름차순, 내림차순 정렬

```python
#인덱스로 오름차순 정렬
df.sort_index()
```

```python
#인덱스로 오름차순 정렬
df.sort_index(ascending=False)
df
```
