# 11.데이터 정렬

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
#특정 열에 대한 오름차순 배열
df.sort_values('키')
```

```python
#특정 열에 대한 내림차순 배열
df.sort_values('키', ascending=False)
```

```python
#여러기준 적용한 내림차순 - 첫번째기준(수학), 두번째기준(영어)
df.sort_values(['수학', '영어'], ascending=False)
```

```python
#여러기준 적용한 여러차순 - 첫번째기준(수학-오름차순), 두번째기준(영어-내림차순)
df.sort_values(['수학', '영어'], ascending=[True, False])
```

```python
#특정 열에 대한 데이터 값의 오름차순 - 일부 열의 데이터
df['키'].sort_values()
```

```python
#특정 열에 대한 오름차순 - 전체 데이터
df.sort_values('키')
```

```python
#Index 기준으로 정렬
df.sort_index()
```
