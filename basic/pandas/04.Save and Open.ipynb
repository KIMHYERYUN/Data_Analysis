# 4.파일 저장 및 열기
DataFrame에서 객체를 excel, csv, txt등 형태의 파일로 저장 및 열기

```python
import pandas as pd
data = {
    '이름' : ['kim', 'hur', 'park', 'jung', 'lee', 'choi', 'han'],
    '학교' : ['덕원여고', '덕원여고', '덕원여고', '화곡고', '광영여고', '덕원여고', '화곡고'],
    '키' : [167, 165, 180, 172, 176, 161, 159],
    '국어' : [85, 90, 70, 89, 100, 61, 50],
    '수학' : [95, 81, 73, 86, 91, 40, 80],
    '영어' : [90, 95, 89, 61, 70, 80, 86],
    '과학' : [88, 79, 60, 95, 86, 54, 75],
    '사회' : [80, 93, 91, 80, 94, 79, 81],
    '특기' : ['python', 'java', '', '', 'java', 'javascript', 'python']
}
df=pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번'])
df.index.name = '지원번호'
df
```

## 저장하기
### csv파일로 저장

```python
df.to_csv('score.csv', encoding='utf-8-sig')
```

```python
df.to_csv('score.csv', encoding='utf-8-sig', index=False)
```

### 텍스트 파일로 저장(.txt)

```python
#tap으로 구분된 텍스트 파일
df.to_csv('score.txt', sep='\t')
```

### 엑셀 파일로 저장(.xlsx)

```python
df.to_excel('score.xlsx')
```

***
## 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
#첫번째 1row 건너뛰어 출력
df = pd.read_csv('score.csv', skiprows=1)
df
```

```python
#1, 3, 5 지정된 갯수만큼 row 건너뛰어 출력
df = pd.read_csv('score.csv', skiprows=[1, 3, 5])
df
```

```python
#지정된 갯수만큼 가져와서 출력하기
df = pd.read_csv('score.csv', nrows=4)
df
```

```python
#지정된 영역에서 갯수만큼 가져와서 출력 
# -2행 제거 후 1행은 제목으로 인지 후 3행 출력
df = pd.read_csv('score.csv', skiprows=2, nrows=3)
df
```

### 텍스트 파일 열기

```python
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
#Index 설정하여 열기
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호')
df
```

```python
#파일열고 Index 설정하고 출력
df = pd.read_csv('score.txt', sep='\t')
df.set_index('지원번호', inplace=True)
df
```

### 엑셀파일 열기

```python
df = pd.read_excel('score.xlsx')
df
```

```python
#Index 설정하고 열기
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```
