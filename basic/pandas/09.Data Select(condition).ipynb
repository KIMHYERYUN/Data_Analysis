# 9.데이터 선택(조건)

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
#조건이 맞는지 여부 확인 - True/False
df['키'] >= 160
```

```python
#조건이 맞는 데이터 가져오기
df[df['키'] >= 160]
```

```python
#조건이 맞는 데이터 가져오기 - 변수활용
filt = (df['키'] >= 160)
df[filt]
```

```python
#조건이 맞지 않은 데이터 가져오기 - 변수활용
df[~filt]
```

```python
#조건에 맞는 데이터 가져오기 - loc 활용
df.loc[df['키'] >= 160, '수학']
```

```python
#조건에 맞는 데이터 가져오기 - loc 활용
df.loc[df['키'] >= 160, ['수학', '과학']]
```

## 다양한 조건

### &(그리고, and)

```python
#여러조건에 모두 해당되는 데이터 가져오기 - loc 활용
df.loc[(df['키'] >= 160) & (df['수학'] > 80)]
```

### |(또는, or)

```python
#여러조건마다 해당되는 데이터 가져오기 - loc 활용
df.loc[(df['키'] < 160) | (df['수학'] > 80)]
```

### str함수
문자열 관련 함수 및 처리

```python
#이름 중 'h'로 시작하는 사람
filt = df['이름'].str.startswith('h')
df[filt]
```

```python
#이름 중 i가 들어간 사람
filt = df['이름'].str.contains('i')
df[filt]
```

```python
#이름 중 i가 들어가지 않은사람
df[~filt]
```

```python
#일부 단어가 들어간 데이터 추출
langs = ['python', 'java']
filt = df['특기'].isin(langs)
df[filt]
```

```python
#일부 단어가 들어간 데이터 추출 - 대소문자 구분없이
langs = ['python', 'java']
filt = df['특기'].str.lower().isin(langs)
df[filt]
```

```python
#결측치(NaN) 제거하여 조건 데이터 가져오기
#-str.contains 까지 True, False로 판별
#-단, NaN을 True, False 설정가능
filt = df['특기'].str.contains('java', na=False)
df[filt]
```

```python
#결측치(NaN)와 같이 조건 데이터 가져오기
filt = df['특기'].str.contains('java', na=True)
df[filt]
```
