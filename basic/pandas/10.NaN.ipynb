## 데이터 채우기 - fillna

```python
#NaN을 빈칸으로 채우기
#-파일에서 불러왔을 때 NaN display 된다
df.fillna('')
```

```python
#NaN을 '없음'으로 채우기
df.fillna('없음')
```

```python
#학교 데이터를 NaN으로 채우기
import numpy as np
df['학교'] = np.nan
df
```

```python
#NaN를 '모름' 채우기 - 단, 동기화 X
df.fillna('모름')
df
```

```python
#NaN를 '모름' 채우기 - 단, 동기화 ㅇ
df.fillna('모름', inplace=True)
df
```

## 데이터 제외하기 - dropna

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
#NaN 포함하는 데이터를 삭제 - 동기화 X
df.dropna()
```

```python
#NaN 포함하는 데이터를 삭제 - 동기화 ㅇ
df.dropna(inplace=True)
df
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

dropna 함수의 속성
- axis : index or columns
- how : any(값이 하나이상) or all(모두 다)

```python
df.dropna(axis='index', how='any')
```

```python
df.dropna(axis='columns', how='any')
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all')
```
