# 4.스타일

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Window
#matplotlib.rcParams['font.family'] = 'AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15 #글자크기 
#한글 폰트 사용시, 마이너스 표시 깨짐 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 5, 10]
```

```python
plt.plot(x, y)
```

```python
#데이터 표시 모양
plt.plot(x, y, marker='o')
```

```python
#데이터 표시 크기
plt.plot(x, y, marker='o', markersize=10)
```

```python
#데이터 표시 테두리 설정
plt.plot(x, y, marker='o', markeredgecolor='red')
```

```python
#데이터 표시 색 설정
plt.plot(x, y, marker='o', markerfacecolor='red')
```

## 선 스타일

```python
#선 없애기 - marker
plt.plot(x, y, marker='o',linestyle='None')
```

```python
#선 모양(1)
plt.plot(x, y, linestyle=':')
```

```python
#선 모양(2)
plt.plot(x, y, linestyle='--')
```

```python
#선 모양(2)
plt.plot(x, y, linestyle='-')
```

## 색깔

```python
plt.plot(x, y, color='red')
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='r')
```

## 간단한 linestyle 적용형식

```python
#color,marker,linestyle 입력
plt.plot(x, y, 'ro--')
```

```python
#color,marker,linestyle 입력 - 필수 x
plt.plot(x, y, 'g--')
```

## 축약어

```python
#간단한 축약어 사용
plt.plot(x, y, marker='o', mfc='green', ms='10', mec='blue', ls='-.')
```

## 투명도

```python
#투명도 - alpha : 0 ~ 1 
plt.plot(x, y, marker='o', mfc='green', ms='10', mec='blue', ls='-.', alpha=0.5)
```

## 그래프 크기

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
#확대 및 축소
plt.figure(figsize=(5, 10), dpi=200)
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#00fa05')
plt.plot(x, y)
```
