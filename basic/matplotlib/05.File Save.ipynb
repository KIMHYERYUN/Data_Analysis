# 5. 파일 저장

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Window
#matplotlib.rcParams['font.family'] = 'AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15 #글자크기 
#한글 폰트 사용시, 마이너스 표시 깨짐 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 5, 10]
```

```python
plt.plot(x, y)
```

```python
#해상도 미적용, 저장 - 기본상태로 저장
plt.plot(x, y)
plt.savefig('graph.png')
```

```python
#해상도 적용하면서 저장 - 높아진 상태로 저장
plt.plot(x, y)
plt.savefig('graph_100.png', dpi=100)
```

```python
#해상도 적용한 뒤 저장 - 높아진 상태로 저장
plt.figure(dpi=200)
plt.plot(x, y)
plt.savefig('graph_200.png')
```

```python
#보여지는 해상도, 저장되는 해상도 = 200, 400
plt.figure(dpi=200)
plt.plot(x, y)
plt.savefig('graph_400.png', dpi = 400)
```
