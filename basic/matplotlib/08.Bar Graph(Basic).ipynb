# 8. 막대그래프(기본)

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Window
#matplotlib.rcParams['font.family'] = 'AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15 #글자크기 
#한글 폰트 사용시, 마이너스 표시 깨짐 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['kim', 'hur', 'park']
values = [180, 165, 171]
plt.bar(labels, values)
```

```python
#색깔 변경
plt.bar(labels, values, color='yellow')
```

```python
#색깔 변경 - 다중
colors = ['r', 'g', 'yellow']
plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
#막대 굵기 설정
plt.bar(labels, values, width=0.5)
```

## 데이터 제한하여 그리기

```python
#각 축에 나타날 수 있는 영역 설정
plt.bar(labels, values)
plt.ylim(155, 190)
```

## 축 이름 데이터 설정 변경

```python
#축의 값 45도로 기울이기
plt.bar(labels, values, width=0.5)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```python
#축의 값 이름 변경
ticks =['1번학생', '2번학생', '3번학생']
plt.bar(labels, values)
plt.xticks(labels, ticks)
```
