# 6. 텍스트

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Window
#matplotlib.rcParams['font.family'] = 'AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15 #글자크기 
#한글 폰트 사용시, 마이너스 표시 깨짐 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 5, 10]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, marker='o')
#y의 값을 하나씩 가져오기 - index와 데이터
for idx, txt in enumerate(y):
    #x좌표, y좌표, 데이터 값
    plt.text(x[idx], y[idx], txt)
```

```python
plt.plot(x, y, marker='o')
#y의 값을 하나씩 가져오기 - index와 데이터
for idx, txt in enumerate(y):
    #x좌표, y좌표 위로 0.3, 데이터 값, 데이터 정렬(horizontal alignment=ha), 색
    plt.text(x[idx], y[idx]+0.3, txt, ha='center', color='red')
```
