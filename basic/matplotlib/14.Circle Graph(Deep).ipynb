# 14.원그래프(심화)

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Window
#matplotlib.rcParams['font.family'] = 'AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15 #글자크기 
#한글 폰트 사용시, 마이너스 표시 깨짐 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values=[30, 25, 20, 13, 10, 2]
labels=['python', 'java', 'javascript', 'java', 'c/c++', 'etc']

#label, percent 넣기('%'는 %%), 시작포인트 지정, 방향설정
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, 
        counterclock=False)
plt.show()
```

```python
#색 변경
values=[30, 25, 20, 13, 10, 2]
labels=['python', 'java', 'javascript', 'java', 'c/c++', 'etc']
explode = [0.2]*6
#colors=['b','g','r','c','m','y']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']

#label, percent 넣기('%'는 %%), 시작포인트 지정, 방향설정
plt.pie(values, labels=labels, colors=colors, explode=explode, 
        autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

## 속성 정의

```python
#도넛 원 모양 - 유의 : {}, []이 아니다
wedgeprops={'width' : 0.2}
plt.pie(values, wedgeprops=wedgeprops, labels=labels, colors=colors, 
        autopct='%.1f%%', startangle=90, counterclock=False)

plt.show()
```

```python
#도넛 원 모양
wedgeprops={'width' : 0.6, 'edgecolor':'white', 'linewidth':5}
plt.pie(values, wedgeprops=wedgeprops, labels=labels, colors=colors, 
        autopct='%.1f%%', startangle=90, counterclock=False)

plt.show()
```

## 일부 데이터만 나타내기

```python
#일부 데이터만 나타낼 수 있는 함수 생성
def custom_autopct(pct):
    return ('%.1f%%' % pct if pct >= 10 else '')

#autopct 부분에 해당 함수 넣기 : '%.1f%%' -> custom_autopct
plt.pie(values, wedgeprops=wedgeprops, labels=labels, colors=colors, 
        autopct=custom_autopct, startangle=90, counterclock=False)
plt.show()
```

```python
#일부 데이터만 나타낼 수 있는 함수 생성
def custom_autopct(pct):
    #return ('%.1f%%' % pct if pct >= 10 else '')
    return '{:.1f}%'.format(pct) if pct >=10 else ''

#autopct 부분에 해당 함수 넣기 : '%.1f%%' -> custom_autopct
plt.pie(values, wedgeprops=wedgeprops, labels=labels, colors=colors, 
        autopct=custom_autopct, startangle=90, counterclock=False)
plt.show()
```

```python
#일부 데이터만 나타낼 수 있는 함수 생성
def custom_autopct(pct):
    #return ('%.1f%%' % pct if pct >= 10 else '')
    return '{:.0f}%'.format(pct) if pct >=10 else ''

#autopct 부분에 해당 함수 넣기 : '%.1f%%' -> custom_autopct
plt.pie(values, wedgeprops=wedgeprops, labels=labels, colors=colors, 
        autopct=custom_autopct, startangle=90, counterclock=False)
plt.show()
```

```python
#퍼센트 글자 위치 변경 : pctdistance 
#일부 데이터만 나타낼 수 있는 함수 생성
def custom_autopct(pct):
    #return ('%.1f%%' % pct if pct >= 10 else '')
    return '{:.0f}%'.format(pct) if pct >=10 else ''

#autopct 부분에 해당 함수 넣기 : '%.1f%%' -> custom_autopct
plt.pie(values, wedgeprops=wedgeprops, labels=labels, colors=colors, 
        autopct=custom_autopct, startangle=90, counterclock=False, pctdistance=0.7)
plt.show()
```
