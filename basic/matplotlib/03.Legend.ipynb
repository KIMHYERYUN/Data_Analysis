# 3. 범례

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Window
#matplotlib.rcParams['font.family'] = 'AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15 #글자크기 
#한글 폰트 사용시, 마이너스 표시 깨짐 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 5, 10]
```

```python
#matplot.plt.legend 옵션 넣기 : legendㅇ
#google 검색
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
plt.plot(x, y, label='데이터명')
plt.legend()
```

```python
#범례 위치조정 - 위치명
plt.plot(x, y, label='데이터명')
plt.legend(loc='lower right')
```

```python
#범례 위치조정 - x,y축 활용(0 ~ 1)
plt.plot(x, y, label='데이터명')
plt.legend(loc=(0.5, 0.7))
```

## legend 표시 형태

```python
idx = [0, 1, 2]
x = [1, 2, 3]
y = [2, 5, 10]
z = [3, 1, 6]

plt.plot(idx, x, label='x')
plt.plot(idx, y, label='y')
plt.plot(idx, z, label='z')

plt.legend(ncol=3)
```
