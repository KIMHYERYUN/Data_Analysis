# 2.축

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Window
#matplotlib.rcParams['font.family'] = 'AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15 #글자크기 
#한글 폰트 사용시, 마이너스 표시 깨짐 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 5, 10]
```

```python
plt.plot(x,y)
#개별 폰트 설정
plt.title('꺽은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20})
```

## 축 라벨링

```python
plt.plot(x, y)
plt.xlabel('x축')
plt.ylabel('y축')
```

```python
#색, 위치([left, center, right], [top, center, bottom]) 변경
plt.plot(x, y)
plt.xlabel('x축', color='red', loc='right')
plt.ylabel('y축', color='#00aa00', loc='top')
```

```python
#축의 값 지정
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.ylabel([3, 6, 9, 12])
```
