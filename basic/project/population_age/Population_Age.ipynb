## 남자 데이터 정의

### 2011년

```python
#남자 - 행정기관 및 연령별 인구 수 추출
#1. 파일 불러오기
#2. 불필요한 열 건너뛰기
#3. 인덱스 설정
#4. 해당 데이터 선택 - B, E~Y열
import pandas as pd
df_m_2012 = pd.read_excel('2012_연령별인구현황_월간.xlsx', skiprows=3,
                    index_col = '행정기관', usecols='B, E:Y')

df_m_2012.head()
```

```python
#전국에 해당하는 연령별 인구수 추출
#1. 문자열로 되어있는 숫자 -> 데이터 유형 변경 : 콤마 삭제, int로 변경 -> 반드시 저장
df_m_2012.iloc[0] = df_m_2012.iloc[0].str.replace(',','').astype(int)
```

```python
df_m_2012.iloc[0]
```

## 남자, 여자 데이터를 가지고 그래프 그리기

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Window
#matplotlib.rcParams['font.family'] = 'AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15 #글자크기 
#한글 폰트 사용시, 마이너스 표시 깨짐 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

### 2012년 남자, 여자 비교

```python
#연령 별 남자 데이터 - 수평 막대 그래프

#그래프 크기 변경
plt.figure(figsize=(10, 7))

#단위 : 천명으로 변환
plt.barh(df_m_2012.columns, df_m_2012.iloc[0] // 1000)
```

```python
#연령 별 여자 데이터 - 수평 막대 그래프

#그래프 크기 변경
plt.figure(figsize=(10, 7))

#단위 : 천명으로 변환
plt.barh(df_w_2012.columns, df_w_2012.iloc[0] // 1000)
```

```python
plt.barh(df_m_2012.columns, df_m_2012.iloc[0] // 1000)
plt.barh(df_w_2012.columns, df_w_2012.iloc[0] // 1000)
plt.show()
```

```python
#대칭적으로 보여주는 그래프 그리기
plt.barh(df_m_2012.columns, -df_m_2012.iloc[0] // 1000)
plt.barh(df_w_2012.columns, df_w_2012.iloc[0] // 1000)

#타이틀
plt.title('2012년 대한민국 인구 피라미드')
#오류 : show 하기 전 저장 필수!!
plt.savefig('2012_인구피라미드.png', dpi=100)
plt.show()
```

### 2022년 남자, 여자 비교
