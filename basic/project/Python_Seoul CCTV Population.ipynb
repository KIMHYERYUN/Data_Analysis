서울시 자치구 년도별 CCTV와 인구관계 분석

```python
import numpy as np
import pandas as pd
import seaborn as sns

from pandas import Series, DataFrame 

#그래프
import matplotlib.pyplot as plt

#시각화할 때 한글을 출력하기 위해서
import platform
from matplotlib import font_manager, rc
#한글처리
#메켄도시의 경우
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
#윈도우의 경우
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname = "c://Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
    
    
#cctv 데이터 읽어오기
cctv = pd.read_excel('c://programming/python/pandasdata/cctv.xlsx')
cctv.info()

#인구 데이터 읽어오기
#pop = pd.read_csv('c://programming/python/pandasdata/pop.txt', encoding='utf-8', delimeter='\t', thousands=',', 
#skiprows=2)
#pop                  
pop = pd.read_csv('c://programming/python/pandasdata/pop.txt', encoding='utf-8', skiprows=2, delimiter='\t',
thousands=',')
pop
```

열이름 변경과 인덱스 설정

```python
#cctv에서 첫번째 컬럼의 이름을 구별로 변경
cctv.rename(columns={cctv.columns[0]:'구별'}, inplace=True)
cctv.info()

#pop에서 두번째 컬럼의 이름을 구별로 변경
pop.rename(columns={pop.columns[1]:'구별'}, inplace=True)
pop.info()
```

#공통된 컬럼의 값을 동일한 구조로 변경

```python
#cctv의 구별 데이터의 공백을 모두 제거
gu = []
for x in cctv['구별']:
    gu.append(x.replace(' ', ''))

cctv['구별'] = gu
print(cctv['구별'])
```

데이터 필터링

```python
#pop에서 기간, 구별, 계, 남자, 여자 열만 추출
pop_filter = pop[['기간', '구별', '계', '남자', '여자']]
pop_filter
                  
#pop_filter에서 첫번째 행 제거
pop_filter_result = pop_filter.drop([0])
pop_filter_result

#새로운 열 추가
pop_filter_result['여성비율'] = pop['여자']/pop['계'] * 100
pop_filter_result        
```

2개의 데이터 합치기 - key는 구별

```python
#2개의 데이터 join - 구별로 합치기
df = pd.merge(cctv, pop_filter_result, on='구별')
df
```

불필요한 열 제거

```python
#지우고 싶은 열과 axis=1의 값은 열을 지우기 위한 옵션
df_filtering = df.drop(['2012년 이전', '2013년', '2014년', '2015년', '2016년', '2017년', '2018년', '기간'], axis=1)

df_filtering
```

###구별을 인덱스로 설정

```python
df_idx = df_filtering.set_index('구별')
df_idx
```

막대 그래프 그리기

```python
#cctv 개수를 시각화
df_idx['총계'].sort_values().plot(kind='barh', grid=True, figsize=(5, 5))
plt.show()
```

```python
#cctv 개수를 인구수로 나눈 결과를 시각화
df_idx['cctv비율'] = df_idx['총계']/df_idx['계']*100
df_idx['cctv비율'].sort_values().plot(kind='barh', grid=True, figsize=(5,5))
plt.show()
```

산점도 그리기

```python
plt.figure(figsize=(5,5))

plt.scatter(df['계'], df['총계'], s=50)

plt.xlabel('인구수')
plt.ylabel('CCTV 개수')

plt.show()
```

```python
산점도에 회귀선 그리기
```

```python

fp1 = np.polyfit(df['계'], df['총계'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

#df_idx['잔차'] = np.abs(df_idx['총계'] - f1(df_idx['계']))
plt.figure(figsize=(10, 10))
plt.scatter(df['계'], df['총계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

오차 표시

```python
fp1 = np.polyfit(df['계'], df['총계'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
df['오차'] = np.abs(df['총계'] - f1(df['계']))
plt.figure(figsize=(14,10))
plt.scatter(df['계'], df['총계'], c=df['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(24):
    plt.text(df['계'][n]*1.02, df['총계'][n]*0.98,
        df.index[n], fontsize=12)

plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
